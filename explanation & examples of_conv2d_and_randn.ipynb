```python
# With square kernels and equal stride
m = nn.Conv2d(1,5,3, stride=2)
input = torch.randn(1, 3, 5)
output = m(input)
print("input=",input)
print("---------------------------------------")
print("output=",output)
```

```python
# With square kernels and equal stride
m = nn.Conv2d(1,2,3, stride=2)
input = torch.randn(1, 3, 5)
output = m(input)
print("input=",input)
print("---------------------------------------")
print("output=",output)
```

```python
# With square kernels and equal stride
m = nn.Conv2d(1,5,(3,2), stride=1)
input = torch.randn(1, 5, 5)
output = m(input)
print("input=",input)
print("---------------------------------------")
print("output=",output)
```

```python
# With square kernels and equal stride
m = nn.Conv2d(2,5,3, stride=2)
input = torch.randn(2, 3, 5)
output = m(input)
print("input=",input)
print("---------------------------------------")
print("output=",output)
```

```python

# non-square kernels and unequal stride and with padding
m = nn.Conv2d(2, 3, (3, 5), stride=(2, 1), padding=(4, 2))
# non-square kernels and unequal stride and with padding and dilation
m = nn.Conv2d(16, 33, (3, 5), stride=(2, 1), padding=(4, 2), dilation=(3, 1))
input = torch.randn(20, 16, 50, 100)
output = m(input)
```
